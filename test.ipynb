```python
from speechspectrum import Client

client = Client(
    openai_api_key="YOUR-OPENAI-API-KEY",
    openai_org_id="YOUR-OPENAI-ORG-ID",
    openai_project_id="YOUR-OPENAI-PROJECT-ID",
)
```

```python
# Live pipeline (requires valid OpenAI credentials and an audio file path)
audio_path = "./recording.m4a"
verbatim = client.audio_to_verbatim(audio_path).text
print(verbatim)

nonverbatim = client.verbatim_to_nonverbatim(verbatim)
print(nonverbatim)

enhanced = client.nonverbatim_to_enhanced(nonverbatim)
print(enhanced)

bulletpoints = client.enhanced_to_bulletpoints(enhanced)
print(bulletpoints)
```
